```python
#Exercicis de Train-Test amb ScikitLearn a partir d'un dataset amb informació sobre les vivendes de Houston.
#- Exercici 1
#Parteix el conjunt de dades adjunt en train i test. Estudia els dos conjunts per separat, a nivell descriptiu.
#També adjunt trobaràs una descripció de les diferents variables del dataset.

import numpy as np
import pandas as pd
df =  pd.read_csv("housing data.csv",encoding='latin-1',delimiter=";")

df.columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

print(df.head())

from sklearn.model_selection import train_test_split 

train, test = train_test_split(df, test_size = 0.30)

print("--> Exemples emprats per entrenar: ", len(train))
print("--> Exemples del test: ", len(test))

# Estudia els dos conjunts per separat, a nivell descriptiu.


print("\n--> Tamany del dataset d'entrenament:\n", train.shape)
print("\n--> Tamany del dataset TEST:\n", test.shape, "\n")

print("\n--> Informació estadística del dataset d'entrenament:\n",train.describe())
print("\n--> Informació estadística del dataset per al test:\n",train.describe())


print("\n--> Informació del dataset d'entrenament:\n")
print(train.info())
print("\n--> Informació del dataset per al test:\n")
print(test.info())

```

```python
#- Exercici 2
#Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...).

#--> M'interessa fer una variable ficticia de l'índex d'accessos a autopistes "index of accessibility to radial highways"

autopistes = pd.get_dummies(df["RAD"])
print("--> Variable ficticia autopistes:\n", autopistes.head(5))

# M'interessa interessa estandarditzar les dades de la contaminació per òxids de nitrogen. 

from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
NOx_values = df[['NOX']]
scaled_values = min_max_scaler.fit(NOx_values)
print("\n--> Normalització NOx:\n", min_max_scaler.transform(NOx_values)[0:5])
```
